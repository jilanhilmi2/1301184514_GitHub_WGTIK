```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk melakukan histogram equalization
def histogram_equalization(image):
    # Ubah citra menjadi grayscale (jika tidak sudah grayscale)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Hitung histogram citra asli
    hist_original = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

    # Equalisasi histogram
    equalized_image = cv2.equalizeHist(gray_image)

    # Hitung histogram citra hasil equalisasi
    hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

    return gray_image, hist_original, equalized_image, hist_equalized
```

```python
# Membaca citra RGB
input_image = cv2.imread('IMG_20180812.jpg')
```

```python
# Melakukan histogram equalization
gray_image, hist_original, equalized_image, hist_equalized = histogram_equalization(input_image)
```

```python
# Tampilkan citra asli, histogram asli, citra hasil equalisasi, dan histogram hasil equalisasi
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.plot(hist_original)
plt.title('Histogram Asli')
plt.grid()

plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Citra Hasil Equalisasi')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(hist_equalized)
plt.title('Histogram Hasil Equalisasi')
plt.grid()

plt.show()
```
